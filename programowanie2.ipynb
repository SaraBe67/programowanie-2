```python
stress = X_train.copy()
stress_labels = y_train.copy()
```

# **Czyszczenie danych**
**potok danych pipeline**

Etap czyszczenia danych polegał na weryfikacji kompletności oraz spójności zbioru. Przeprowadzona analiza wykazała brak wartości pustych oraz brak zduplikowanych rekordów.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Wybieramy wszystkie kolumny oprócz zmiennej objaśnianej (stress_level)
num_features = X_train.columns.tolist()

# Dane numeryczne i nie ma braków -> ColumnTransformer aplikuje wyłącznie StandardScaler
# StandardScaler sprowadza wszystkie liczby do wspólnej, podobnej skali, nie zmieniając ich znaczenia
full_pipeline = ColumnTransformer([
    ('num', StandardScaler(), num_features)])

# Dopasowywanie i przekształcenie danych treningowych w macierz liczb
X_train_prepared = full_pipeline.fit_transform(X_train)

X_train_prepared.shape
```

```python
pd.DataFrame(X_train_prepared, columns=stress.columns)
```

# **Regresja Liniowa**
**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_train_prepared, stress_labels)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
stress_predictions1 = lin_reg.predict(X_train_prepared)
lin_mse = mean_squared_error(stress_labels, stress_predictions1)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
def display_scores(scores):
 print("Scores:", scores)
 print("Mean:", scores.mean())
 print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score

lin_scores = cross_val_score(lin_reg, X_train_prepared, stress_labels, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

# **Regresja lasów losowych**
**Random Forest Regressor Model**

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(X_train_prepared, stress_labels)

stress_predictions2 = forest_reg.predict(X_train_prepared)
forest_mse = mean_squared_error(stress_labels, stress_predictions2)
forest_rmse = np.sqrt(forest_mse)

forest_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(forest_reg, X_train_prepared, stress_labels,
scoring="neg_mean_squared_error", cv=3)
forest_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(forest_rmse_scores)
```

# **Grid Search**

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [30, 50, 100],
        'max_features': [2, 4, 6, 'sqrt'],
        'max_depth': [None, 10, 20]}]

forest_reg = RandomForestRegressor(random_state=42)

grid_search = GridSearchCV(
    forest_reg,
    param_grid,
    cv=3,
    scoring='neg_mean_squared_error',
    return_train_score=True,
    n_jobs=-1
)

grid_search.fit(X_train_prepared, stress_labels)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

## poprawka

```python
best_forest_model = grid_search.best_estimator_

forest_predictions = best_forest_model.predict(X_train_prepared)

forest_mse = mean_squared_error(stress_labels, forest_predictions)
forest_rmse = np.sqrt(forest_mse)

print(forest_rmse)
```

# **Ocena na zbiorze testowym**

```python
best_model = grid_search.best_estimator_
```

```python
X_test_prepared = full_pipeline.transform(X_test)
```

```python
test_pred = best_model.predict(X_test_prepared)
```

```python
test_estim = mean_squared_error(y_test, test_pred)
```

```python
final_rmse = np.sqrt(test_estim)
final_rmse
```
